```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("Datafiles/data-1.xlsx")
```

```python
df
```

```python
df["Date"]=pd.to_datetime(df["Date"])  #convert to datetime format
```

```python
#plotting the data
```

```python
plt.plot(df["Date"],df["Sales"],color="green")
plt.title("Sales over time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
#creating a bar chart
plt.bar(df["Product"],df["Sales"],color="green")
plt.title("Sales of products")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```
