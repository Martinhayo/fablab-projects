```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("car data.xlsx")
```

```python
df.head()
```

```python
variables=['Min Price (Lakh)','Max Price (Lakh)','Range (kmpl)','CC','Seats','Variants','Ex-Showroom Price','RTO','Insurance','Other','Onroad Price']
```

```python
plt.figure(figsize=(15,12))
for i,var in enumerate(variables,start=1):
    plt.subplot(4,3,i)
    sns.scatterplot(x=df[var],y=df['Max Price (Lakh)'],alpha=0.7)
    plt.title(f'Scatter plot of {var} vs Onroad Price')
    plt.xlabel(var)
    plt.ylabel('Target')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,12))
for i,var in enumerate(variables,start=1):
    plt.subplot(3,4,i)
    sns.histplot(df[var],bins=20,kde=True,color="green")
    plt.title(f'histogram of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#creating a box plot
data=[12,15,14,10,9,8,7,19,21,25,26,28,50,100]
sns.boxplot(data=data)
plt.show()
```

```python
#creat a bboxplot of Range by Car type
sns.boxplot(data=df,x="Type",y="Range (kmpl)",hue='Type',palette=['green','red'])
plt.title('Boxplot of Range by car type')
plt.show()
```

```python
#heatmap
selectedColumns=df[['Min Price (Lakh)','Max Price (Lakh)','Range (kmpl)','CC','Seats',
                 'Variants','Ex-Showroom Price','RTO','Insurance','Other']]
```

```python
selectedColumns.head()
```

```python
corr=selectedColumns.corr()     #showing correlation between the variables
corr
```

```python
#plotting heatmap
sns.heatmap(corr,annot=True,cmap="Reds")
plt.title("heatmap of correlation matrix")
plt.show()
```

```python
#violin plot of range by car type
sns.violinplot(data=df,x='Type',y='Range (kmpl)',hue='Type',palette=["green","red"])
plt.title('violin plot of range by car type')
plt.show()
```

```python
sns.violinplot(data=df,x='Type',y='Max Price (Lakh)',hue='Type',palette=["green","red"])
plt.title('violin plot of max prie by car type')
plt.show()
```
